```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor, plot_tree
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import sklearn.datasets
import sklearn.metrics as metrics
```

Importing the Boston House Price Dataset

```python
house_price_dataset = sklearn.datasets.fetch_california_housing()
```

```python
print(house_price_dataset)
```

```python
# Loading the dataset to a pandas dataframe
df = pd.DataFrame(house_price_dataset.data, columns = house_price_dataset.feature_names)
```

```python
df.head()
```

```python
# add the target column to the dataframe
df['price'] = house_price_dataset.target
```

```python
df.head()
```

```python
# checking the number of rows and columns in the dataframe
df.shape
```

```python
# check for missing values
df.isnull().sum
```

```python
# statistical measures of the dataset
df.describe()
```

Understanding the **correlation** between various features in the dataset

1. Positive Correlation
2. Negative Correlation

```python
correlation = df.corr()
```

```python
# constructing a heatmap to understand the correlation

plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

Splitting the data and target

```python
# Project Modules
## Module 1: Data Preprocessing
print("Module 1: Data Preprocessing")
df.fillna(df.median(), inplace=True)
```

```python
## Module 2: Feature Selection
print("Module 2: Feature Selection")
X = df.drop(columns=['price'])
Y = df['price']
```

```python
print(X,Y)
```

Splitting the data into training data and test data

```python
# Splitting dataset
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=42)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

XGBoost Regressor

```python
# load the model
model = XGBRegressor()
```

```python
#training the model with X_train
model.fit(X_train, Y_train)
```

Evaluation

Prediction on training data

```python
# accuracy for prediction on training data
training_data_prediction = model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
# R Squared Error
score_1 = metrics.r2_score(Y_train, training_data_prediction)

# Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_train, training_data_prediction)

print('R Sqaured Error:', score_1)
print('Mean Absolute Error:', score_2)
```

Visualize the actuale prices and predicted prices

Prediction on test data

```python
# accuracy for prediction on test data
test_data_prediction = model.predict(X_test)
```

```python
# R Squared Error
score_1 = metrics.r2_score(Y_test, test_data_prediction)

# Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_test, test_data_prediction)

print('R Sqaured Error:', score_1)
print('Mean Absolute Error:', score_2)
```

```python
# Scaling data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
## Module 3: Model Training
print("Module 3: Model Training")
models = {
    'Linear Regression': LinearRegression(),
    'Decision Tree': DecisionTreeRegressor(),
    'SVR': SVR(kernel='rbf'),
    'ANN': MLPRegressor(hidden_layer_sizes=(100, 100), max_iter=500, random_state=42),
    'XGBoost': XGBRegressor(n_estimators=100)
}
```

```python
# Train & Evaluate Models
results = {}
for name, model in models.items():
    model.fit(X_train, Y_train)
    Y_pred = model.predict(X_test)
    results[name] = {
        'MAE': mean_absolute_error(Y_test, Y_pred),
        'MSE': mean_squared_error(Y_test, Y_pred),
        'RMSE': np.sqrt(mean_squared_error(Y_test, Y_pred)),
        'R2 Score': r2_score(Y_test, Y_pred)
    }
```

```python
## Module 4: Model Evaluation & Comparison
print("Module 4: Model Evaluation & Comparison")
results_df = pd.DataFrame(results).T
print(results_df)
```

```python
# Visualization
plt.figure(figsize=(10, 5))
sns.barplot(x=results_df.index, y=results_df['R2 Score'], palette='viridis')
plt.xlabel('Model')
plt.ylabel('R2 Score')
plt.title('Model Performance Comparison')
plt.xticks(rotation=45)
plt.show()
```

```python
# Linear Regression: Actual vs. Predicted
plt.figure(figsize=(8, 6))
plt.scatter(Y_test, models['Linear Regression'].predict(X_test))
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Linear Regression: Actual vs. Predicted")
plt.show()
```

```python
# Decision Tree: Tree Visualization
plt.figure(figsize=(20, 10))
plot_tree(models['Decision Tree'], filled=True, feature_names=house_price_dataset.feature_names, rounded=True)
plt.title("Decision Tree Visualization")
plt.show()
```

```python
# SVR: Actual vs. Predicted
plt.figure(figsize=(8, 6))
plt.scatter(Y_test, models['SVR'].predict(X_test))
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("SVR: Actual vs. Predicted")
plt.show()
```

```python
# ANN: Actual vs. Predicted
plt.figure(figsize=(8, 6))
plt.scatter(Y_test, models['ANN'].predict(X_test))
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("ANN: Actual vs. Predicted")
plt.show()
```

```python
# XGBoost: Feature Importance
plt.figure(figsize=(10, 6))
feature_importance = models['XGBoost'].feature_importances_
sns.barplot(x=feature_importance, y=house_price_dataset.feature_names)
plt.xlabel("Feature Importance")
plt.ylabel("Features")
plt.title("XGBoost: Feature Importance")
plt.show()
```
